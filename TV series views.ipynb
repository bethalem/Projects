```python
#Import libraries
import pandas as pd
import numpy as np
```

#### Step 1: Using the DATA sheet in this workbook, read DATA into a dataframe in Python.

```python
# Assign path of excel file
file_path = "./Internship Test - Part 1 and 2 - Python Tableau.xls"
# Read the excel sheet to pandas dataframe
df = pd.read_excel(file_path, sheet_name="DATA")
```

```python
# Show the fitst 10 rows of the dataframe
df.head(10)
```

#### Step 2 : Add a column `siteTypeStarts` that holds the sum of starts of the series and site type. For example, the first value in this column should contain total starts for Blindspot on MVPD.

```python
# This calculates the sum of starts of the series and transforms the groupby() result
df["siteTypeStarts"] = df.groupby(by =["site_type", "series"])["starts"].transform('sum')
df
```

#### Step 3: Add a column 'nameOfDay' that holds the name of the day of that week (for example, Friday).

```python
# This gives the weekday names
df["nameOfDay"] = pd.to_datetime(df["eventdate"]).dt.day_name()
df
```

#### Step 4: Add a column 'startOfWeek' that holds the Monday date of that week. Assume each week starts on Monday.

```python
#This sets 'startOfWeek' to be the first Monday of the week
df["startOfWeek"] = df['eventdate'].dt.to_period('W').apply(lambda r: r.start_time)
df
```

#### Step 5: Write the data frame, not including index numbers, to a comma-separated file named Answers.csv

```python
df.to_csv("Answers.csv", index = False)
```
